# BTVN thú vị

# cho ma trận = [[99,99,99],
                [99,99,99],
                [99,99,99]]

# hãy tưởng tượng ô 99 trống 
# giả sử  0 là o
# giả sử 1 là x
# lấy đầu vào từ phía x và phí o luôn phiên 
# Nếu đầu vào từ phía X là (0,0)
# ma trận sẽ trở thành 

# [x,99,99]
# [99,99,99]
# [99,99,99]
# nếu đầu vào từ phía o cũng là (0,0)
# thông báo nhập lại 
# nếu không thì điền vào ma trận

# thử thách: nếu người đầu tiên có 3 phần tử liên tiếp thì dùng trò chơi


```python
a = [[99,99,99],[99,99,99],[99,99,99]]

def win(p):
    for i in range(3):
        if a[i][0]==a[i][1]==a[i][2]==p: return True
        if a[0][i]==a[1][i]==a[2][i]==p: return True
    if a[0][0]==a[1][1]==a[2][2]==p: return True
    if a[0][2]==a[1][1]==a[2][0]==p: return True
    return False

turn = 1  # 1 = X, 0 = O
while True:
    print(a)

    s = input(("X" if turn==1 else "O") + " nhập (row col): ").strip()
    s = s.replace("(", "").replace(")", "").replace(",", " ")
    parts = s.split()

    if len(parts) != 2:
        print("Nhập lại! Ví dụ: 0 0 hoặc 0,0")
        continue

    try:
        r, c = map(int, parts)
    except:
        print("Nhập lại! Chỉ nhập số.")
        continue

    if r < 0 or r > 2 or c < 0 or c > 2:
        print("Nhập lại! row, col chỉ từ 0 đến 2.")
        continue

    if a[r][c] != 99:
        print("Nhập lại! Ô đã có người đánh.")
        continue

    a[r][c] = turn

    if win(turn):
        print(a)
        print(("X" if turn==1 else "O"), "THẮNG!")
        break

    turn = 1 - turn
```

# tạo 1 mảng numpy có kích thức 150 x 5. giả sử mảng này chứa 150 mẫu dữ liệu 
# của sinh vienen  bao gồm chiều cao cân nặng tuổi , luonmgw và điểm trung bình
# tách mảng dựa trên 4 cột đầu tiên thành 1 biến có tên là x và cột cuối cùng thành y
# tách mảng X thành X train X test trong đó X train chứa 70% dữ liệu và tahcs y thành y train và y test trong đó y train chứa 70% dữ leieuj
# tạo 10 tập dữ liệu không chồng chéo nhau

```python
import numpy as np

np.random.seed(0)

data = np.zeros((150, 5))

# chiều cao, cân nặng, tuổi, lương, điểm TB
data[:, 0] = np.random.randint(150, 191, 150)              # cao (cm)
data[:, 1] = np.random.randint(45, 91, 150)                # nặng (kg)
data[:, 2] = np.random.randint(18, 26, 150)                # tuổi
data[:, 3] = np.random.randint(3, 51, 150)                 # lương (triệu)
data[:, 4] = np.round(np.random.uniform(0, 10, 150), 2)    # điểm TB

print(data.shape)
print(data[:5])
```

```python
X = data[:, :4]
y = data[:, 4]

print("X:", X.shape)
print("y:", y.shape)
```

```python
idx = np.random.permutation(150)
train_size = int(0.7 * 150)

train_idx = idx[:train_size]
test_idx  = idx[train_size:]

X_train = X[train_idx]
X_test  = X[test_idx]

y_train = y[train_idx]
y_test  = y[test_idx]

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
idx2 = np.random.permutation(150)
folds = np.array_split(idx2, 10)   # 10 tập, mỗi tập 15 mẫu

print("Số tập:", len(folds))
print("Số phần tử mỗi tập:", [len(f) for f in folds])
```

```python
X_fold1 = X[folds[0]]
y_fold1 = y[folds[0]]

print("X_fold1:", X_fold1.shape)
print("y_fold1:", y_fold1.shape)
```
